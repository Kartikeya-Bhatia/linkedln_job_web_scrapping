```python
df=pd.read_csv("/drive/My Drive/Datasets/final data.csv")
df
```

```python
df["job_description"]=df["job_description"].astype(str).str.lower()
df
```

```python
from nltk.tokenize import RegexpTokenizer
```

```python
regexp=RegexpTokenizer('\w+')
```

```python
df["skill_token"]=df["job_description"].apply(regexp.tokenize)
df.head(3)
```

```python
import nltk
```

```python
nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords
stopwords=nltk.corpus.stopwords.words('english')
print(stopwords)
```

```python
df['skill_token']=df['skill_token'].apply(lambda x:[item for item in x if item not in stopwords])
df.head(3)
```

```python
df["skill_string"]=df["skill_token"].apply(lambda x: ' '.join([item for item in x if len(item)>3]))
df[['job_description','skill_token','skill_string']].head(3)
```

```python
all_words=" ".join([word for word in df["skill_string"]])
```

```python
nltk.download("punkt")
```

```python
tokenized_word=nltk.tokenize.word_tokenize(all_words)
```

```python
from nltk.probability import FreqDist
fdist = FreqDist(tokenized_word)
fdist
```
